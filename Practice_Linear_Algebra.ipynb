Vector

```python
import numpy as np

# Create a vector
v = np.array([2, 4, 6])
print("Vector v:", v)

# Vector operations
print("Magnitude:", np.linalg.norm(v))
print("Dot product with [1, 0, 1]:", np.dot(v, [1, 0, 1]))
```

Matrix Operattion

```python
# Create a matrix
A = np.array([[1, 2], [3, 4]])
print("Matrix A:\n", A)

# Transpose
print("Transpose:\n", A.T)

# Matrix multiplication
B = np.array([[2, 0], [1, 2]])
print("A × B:\n", np.dot(A, B))

# Determinant
print("Determinant of A:", np.linalg.det(A))

# Inverse
print("Inverse of A:\n", np.linalg.inv(A))
```

```python
#dot product and cross product
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define two 3D vectors
a = np.array([2, 3, 1])
b = np.array([1, -2, 4])

# Dot Product (scalar)
dot_product = np.dot(a, b)

# Cross Product (vector)
cross_product = np.cross(a, b)

# 3D plot setup
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors a and b from origin
origin = [0, 0, 0]
ax.quiver(*origin, *a, color='r', label='Vector a')
ax.quiver(*origin, *b, color='b', label='Vector b')

# Plot cross product from origin
ax.quiver(*origin, *cross_product, color='g', label='a × b (cross product)')

# Annotate dot product
ax.text(0.5, 0.5, 0.5, f'Dot Product = {dot_product}', fontsize=10)

# Labels and limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('3D Dot & Cross Product Visualization')
plt.show()

```

function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the exponential function
def exponential(x, a=1, b=1):
    return a * np.exp(b * x)

# Generate x values
x = np.linspace(-2, 2, 100)
y = exponential(x, a=1, b=1)

# Plot the function
plt.plot(x, y, label='y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a grid of x values
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Compute sin(x) * y^2 - x = 0
Z = np.sin(X) * Y**2 - X

# Plot the contour where Z = 0 (i.e., sin(x) * y^2 = x)
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=[0], colors='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Graph of $\sin(x) y^2 = x$')
plt.grid(True)
plt.show()
```

linear transformation

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
from PIL import Image
import io

# Assuming you uploaded an image and its filename is stored in the 'uploaded' dictionary
# Get the filename of the first uploaded file
filename = next(iter(uploaded))

# Open the image from the uploaded bytes
img = Image.open(io.BytesIO(uploaded[filename]))

# Define a scaling factor (e.g., 2 for doubling the size)
scale_factor = 2

# Calculate the new size
new_size = (img.width * scale_factor, img.height * scale_factor)

# Apply the scaling transformation (linear transformation)
enlarged_img = img.resize(new_size, Image.NEAREST)  # NEAREST preserves hard edges

# Save the enlarged image
enlarged_img.save(f'enlarged_{filename}')

print(f"Image enlarged from {img.size} to {new_size} using linear transformation.")
print(f"Saved as 'enlarged_{filename}'")
```

Calculus

```python
import sympy as sp

# Define the variable
x = sp.symbols('x')

# Define the function
f = 2*x*x + 1

# Calculate the derivative
derivative = sp.diff(f, x)

print("The derivative of 2*x*x + 1 with respect to x is:", derivative)
```
